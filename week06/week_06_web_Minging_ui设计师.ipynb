```python
# B-4 建构 参数模板 及 字典_compTag
def parse_url_qs_for_industries (url):
    six_parts = urlparse(url) 
    out = parse_qs(six_parts.query)
    return (out)

# parse_url_qs_for_compTag(list(公司数据选择器链结.values())[0])['compTag']
参数模板 = parse_url_qs_for_industries(list(行业.values())[0])
print(参数模板)
# [ parse_url_qs_for_compTag(x)['compTag'] for x in 公司数据选择器链结.values()]
[ parse_url_qs_for_industries(x)['industries'][0] for x in 行业.values()]

字典_industries = { k:parse_url_qs_for_industries(v)['industries'][0] for k,v in 行业数据选择器链结.items()}
print (字典_industries)
```

```python
# B-5 建构 参数模板  
def 参数模板生成(industries , keyword ):
    参数 = 参数模板.copy()
    参数['industries'] = industries
    参数['key'] = keyword
    return (参数)

参数_industries_UI设计师 = { k:参数模板生成(industries = [v], keyword = ['UI设计师']) for k,v in 字典_industries.items()}
print(参数_industries_UI设计师)
```
